```python
import pandas as pd
import os
import re
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#load the names of the csv files in the current directory
csv_files = [file for file in os.listdir() if file.endswith('.csv')]

for file in csv_files:
    print(file)
```

```python
def clean_year_column(df):
    # Check if 'Year' column exists
    if 'Year' in df.columns:
        # Use regex to extract the starting year from the format "2014-15"
        df['Year'] = df['Year'].apply(lambda x: re.match(r'(\d{4})', str(x)).group(1) if pd.notnull(x) else x)
    return df
def clean_district_column(df):
    # Check if 'District' column exists
    if 'District' in df.columns:
        # Convert the 'District' column to uppercase
        df['District'] = df['District'].str.upper()
    return df
def load_and_clean_data(file_path):
    # Load the dataset
    df = pd.read_csv(file_path)
    
    # Clean the 'Year' column
    df = clean_year_column(df)
    
    # Clean the 'District' column
    df = clean_district_column(df)
    
    return df

#I want to run the above function in loop for all the csv files in the current directory and make changes in the original files and save in the same name in same directory
for file in csv_files:
    cleaned_df = load_and_clean_data(file)
    cleaned_df.to_csv(file, index=False)
    
```

```python
# Now I want to make a csv file which checks the uniique common districts in all the csv files and make a new csv file with the name "common_districts.csv" and save it in the same directory
# I will use set intersection to find the common districts in all the csv files
common_districts = None
for file in csv_files:
    df = pd.read_csv(file)
    if 'District' in df.columns:
        districts = set(df['District'].unique())
        if common_districts is None:
            common_districts = districts
        else:
            common_districts = common_districts.intersection(districts)

# Save the common districts to a new CSV file
if common_districts is not None:
    pd.DataFrame({'District': list(common_districts)}).to_csv('common_districts.csv', index=False)
    
```
