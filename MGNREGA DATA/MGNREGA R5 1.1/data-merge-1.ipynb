```python
##in A Main folder I have year-wise folder ("2014-15", "2015-16", "2016-17", "2017-18", "2018-19", "2019-20", "2020-21", "2021-22", "2022-23") and in each year-wise folder I have few csv files and I want to merge all the csv files into one csv file for each year-wise folder and save it in a new folder named "Merged" in the Main folder.
#from each csv before merging, remove the second row (the row starts with "district"), but the header row also starts with "district", so keep the header row and remove only the second row.
import os
import pandas as pd
from glob import glob
# Define the main folder path
main_folder = 'MGNREGA_DATA'  # Replace with your main folder path

# Define the merged folder path
merged_folder = os.path.join(main_folder, 'Merged')
# Create the merged folder if it doesn't exist
os.makedirs(merged_folder, exist_ok=True)
# Get a list of year-wise folders in the main folder
year_folders = [f for f in os.listdir(main_folder) if os.path.isdir(os.path.join(main_folder, f))]
# Loop through each year-wise folder
for year_folder in year_folders:
    # Get a list of CSV files in the current year-wise folder
    csv_files = glob(os.path.join(main_folder, year_folder, "*.csv"))
    # If there are CSV files in the folder
    if csv_files:
        # Read and merge all CSV files in the current year-wise folder
        merged_df = pd.DataFrame()
        for csv_file in csv_files:
            df = pd.read_csv(csv_file)
            # Remove the second row (the row starts with "district"), but keep the header row
            df = df.drop(df.index[1])  # Remove second row (index 1)
            merged_df = pd.concat([merged_df, df], ignore_index=True)
        # Save the merged DataFrame to a new CSV file in the merged folder
        output_file = os.path.join(merged_folder, f"{year_folder}.csv")
        merged_df.to_csv(output_file, index=False)
        print(f"Merged CSV for {year_folder} saved to {output_file}")
    else:
        print(f"No CSV files found in {year_folder}. Skipping.")

```
