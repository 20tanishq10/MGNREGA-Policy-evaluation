# Pillar 3: Income Effect (Nominal)

Model:
income_it = beta1*mgnrega_it + rainfall + population + FE + epsilon

Primary metric: beta1

```python
from pathlib import Path
import pandas as pd
from analysis_utils import load_and_preprocess, fit_fe_clustered, minmax_normalize, compute_vif, summarize_model

DATA_PATH = Path('Panel_Data 2014-24.csv')
OUT_DIR = Path('outputs')
OUT_DIR.mkdir(exist_ok=True)

df = load_and_preprocess(str(DATA_PATH))
```

```python
formula = 'income ~ mgnrega + rainfall + population + C(region_id) + C(year)'
results, used_df = fit_fe_clustered(formula, df)
print(results.summary())

coef_table = summarize_model(results, ['mgnrega', 'rainfall', 'population'])
coef_table.to_csv(OUT_DIR / 'pillar3_coefficients.csv', index=False)
print(coef_table)

vif = compute_vif(used_df, ['mgnrega', 'rainfall', 'population'])
vif.to_csv(OUT_DIR / 'pillar3_vif.csv', index=False)
print(vif)

b1 = results.params.get('mgnrega', 0.0)
region_raw = used_df.groupby('region_id')['mgnrega'].mean() * b1
pillar3_score = minmax_normalize(region_raw).rename('pillar3_income_score').reset_index()
pillar3_score.to_csv(OUT_DIR / 'pillar3_income_score.csv', index=False)
pillar3_score.head()
```
